## README

In this script, we plot the arena raw figures before cropping them... We use this to visualize what holes are targets...

We then use the targets visualized here and mark them in the photoshop mark images, in order to extract them more precisely using the crop_arena_pictures notebook

```python
import os
import numpy
import modules.plot_func as pltt
#import modules.traj_analysis as tran
#import modules.edge_detector as edged
#import modules.helper_func_class as misc
#import modules.traj_to_step_matrix as tstep
#import modules.process_mouse_trials_lib as plib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def get_targetA_coords(entrance):
    target_coords = (numpy.nan,numpy.nan)
    if entrance == u'SW':
        target_coords = 39.68, 11.97
    if entrance == u'SE':
        target_coords = -11.18, 38.58
    if entrance == u'NE':
        target_coords = -37.48, -11.65
    if entrance == u'NW':
        target_coords = 12.28, -38.74
    return numpy.asarray(target_coords)

def get_targetB_coords(entrance):
    target_coords = (numpy.nan,numpy.nan)
    if entrance == u'SW':
        target_coords = -22.36, -17.80
    if entrance == u'SE':
        target_coords = 18.27, -22.99
    if entrance == u'NE':
        target_coords = 23.46, 18.27
    if entrance == u'NW':
        target_coords = -17.32, 22.83
    return numpy.asarray(target_coords)


all_entrances    = ['SW','SE','NE','NW']
picture_file     = 'arena_picture/2022/BKGDimage-220812.png'
arena_pic        = mpimg.imread(picture_file)
arena_pic_extent = [-151.65, 151.34, -84.88, 85.51] # from Kelly's docs or scripts

fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(12,8))
ax.imshow(arena_pic, extent=arena_pic_extent)

for entrance in all_entrances:
    rA = get_targetA_coords(entrance)
    rR = get_targetB_coords(entrance)
    pltt.plot_point(rA,entrance + ' A', color='r',pointArgs=dict(fillstyle='none',markeredgewidth=2),labelcolor='b')
    pltt.plot_point(rR,entrance + ' B', color='r',pointArgs=dict(fillstyle='none',markeredgewidth=2),labelcolor='b')

ax.set_title(picture_file)

output_picture_file = os.path.join(os.path.dirname(picture_file),os.path.basename(picture_file).replace('.png','_target_locations.png'))
plt.savefig(output_picture_file,format='png',dpi=300,facecolor=(1,1,1,1),bbox_inches='tight')

plt.show()
```
